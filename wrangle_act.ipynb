##  Data Cleaning

```python
# 1-2 converting timestamp type from string to datetime and remove the hour Zeros
DF_TA_clean['timestamp'] = pd.to_datetime(DF_TA_clean.timestamp).dt.date
```

```python
#Test
DF_TA_clean.head(1)
```

```python
# 3-4 removing excess data from Twitter Archive and Tweet_data
DF_TA_clean = DF_TA_clean.loc[:len(DF_IP_clean['tweet_id']),:]
DF_TD_clean = DF_TD_clean.loc[:len(DF_IP_clean['tweet_id']),:]
```

```python
#Test
DF_TA_clean
```

```python
#Test
DF_TD_clean
```

```python
# 5- Remove letters from  name column in Twitter Archive Table
Invalid_Names = DF_TA_clean.name.str.contains('^[a-z]', regex = True)
DF_TA_clean.loc[Invalid_Names,'name'] = 'None'
```

```python
# 6- Remove retweet entries from Twitter_Archive Table
DF_TA_clean = DF_TA_clean[DF_TA_clean.retweeted_status_id.isna()]
```

```python
DF_TA_clean
```

```python
# 7- Remove reply entries from Twitter_Archive
DF_TA_clean = DF_TA_clean[DF_TA_clean.in_reply_to_status_id.isna()]
```

```python
#Test
DF_TA_clean.groupby('retweeted_status_id').sum()
```

```python
#Test
DF_TA_clean.groupby('in_reply_to_status_id').sum()
```

```python
# 8- Rename the p1,p2,p3 columns labels
DF_IP_clean = DF_IP_clean.rename(columns={'p1':'Algorithm_Fisrt_Prediction','p2':'Algorithm_Second_Prediction','p3':'Algorithm_Third_Prediction'})
```

```python
#Test
DF_IP_clean
```

```python
# 9- Rename column id name in tweet_data table
DF_TD_clean = DF_TD_clean.rename(columns={'id':'tweet_id'})
```

```python
#Test
DF_TD_clean
```

```python
# 1- Removing unnecessary columns from Twitter Archive table
DF_TA_clean = DF_TA_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'], axis=1)
```

```python
# 2- Removing  unnecessary columns form tweet_data 
DF_TD_clean = DF_TD_clean.loc[:,['tweet_id','retweet_count','favorite_count']]
DF_TD_clean
```

```python
twitter_archive_master = pd.merge(DF_TA_clean, DF_TD_clean, how='inner' , on='tweet_id')
```
