##### Second insight, I want to know if there is a correlation between favorite count and retweet count.

```python
plt.subplots(figsize=(10, 10))
plt.scatter(DF.retweet_count,DF.favorite_count)
plt.title('Retweets vs Favorites')
plt.xlabel('Retweet Count')
plt.ylabel('Favorite Count');

```

People tend to retweet the tweets they have favorited, which is common sense. Additionally, this is not a normal distribution of data. Most of the tweets counts reside below a threshold of both parameters.

##### Third insight, I want to know what is the relationship between retweet count and time

```python
plt.subplots(figsize=(10, 10))
plt.plot(DF.timestamp, DF.retweet_count)
plt.title('Retweets over time')
plt.xlabel('Date')
plt.ylabel('Retweet Count');
```

I think retweets went through a small decreasing trend in the retweet counts over time. There is a small proportion of tweets that received a large retweet count, whereas the majority remains at a relatively constant level below 10000.

##### Fourth insight, I want to know what is the relation ship between favorite count and time. 

```python
plt.subplots(figsize=(10, 10))
plt.plot(DF.timestamp, DF.favorite_count)
plt.title('Favorites over time')
plt.xlabel('Date')
plt.ylabel('Favorite Count');
```

I think it is similar to what happened with the retweet count. Main points: relatively small decreasing trend and a small proportion of tweets recieved large favorite count.
